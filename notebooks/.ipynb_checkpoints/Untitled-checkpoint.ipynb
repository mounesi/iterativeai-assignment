download kaggle dataset:

```python
import os
import zipfile
from kaggle.api.kaggle_api_extended import KaggleApi

# Create a Kaggle API object
api = KaggleApi()
api.authenticate()

# Download the dataset
dataset_name = 'oguze3s/sign-language-mnist-cnn-keras'
dataset_path = os.path.join(os.getcwd(), 'dataset.zip')
api.dataset_download_files(dataset_name, path=dataset_path, unzip=False)

# Extract the dataset
with zipfile.ZipFile(dataset_path, 'r') as zip_ref:
    zip_ref.extractall(os.getcwd())

# Delete the zip file
os.remove(dataset_path)
```
